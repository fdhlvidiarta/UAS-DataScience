```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Konfigurasi visual
sns.set(style="whitegrid")

# Folder untuk menyimpan hasil visualisasi
os.makedirs("../images", exist_ok=True)

# Nama kolom sesuai dataset SERVO
kolom = ["motor", "screw", "pgain", "vgain", "kelas"]

# Load dataset
df = pd.read_csv("../data/servo.data", header=None, names=kolom)

# Tampilkan 5 data teratas
df.head()
```

```python
# Informasi umum dataset
df.info()
```

```python
# Statistik deskriptif untuk fitur numerik
df.describe()
```

```python
plt.figure(figsize=(7, 4))
sns.histplot(df["kelas"], kde=True)
plt.title("Distribusi Nilai Target (Kelas)")
plt.xlabel("Nilai Kelas")
plt.ylabel("Jumlah Data")
plt.tight_layout()
plt.savefig("../images/eda_distribusi_target.png")
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.boxplot(data=df[["pgain", "vgain"]])
plt.title("Boxplot Fitur Numerik")
plt.ylabel("Nilai")
plt.tight_layout()
plt.savefig("../images/eda_boxplot_fitur.png")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(
    df[["pgain", "vgain", "kelas"]].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Korelasi Antar Variabel Numerik")
plt.tight_layout()
plt.savefig("../images/eda_korelasi.png")
plt.show()
```
